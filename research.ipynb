```python
import ccxt
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
binance=ccxt.binance({
            "apiKey":'apiKey',
            "secret":'apiKey',
        })
symbol='DOGEUSDT'
tf='1h'
```

```python
ratio=binance.fapiDataGetTopLongShortPositionRatio({'symbol':symbol,'period':tf})
ratio
```

```python
timestamp=int(ratio[0]['timestamp'])
df=binance.fetch_ohlcv(symbol,timeframe=tf,since=timestamp)
df=pd.DataFrame(df)
df.columns=['timestamp','open','high','low','close','volume']
df['timestamp']=[datetime.fromtimestamp((x/1000)) for x in df['timestamp']]
df=df.set_index('timestamp')
```

```python
df['ratio']=[float(i['longShortRatio']) for i in ratio]
```

```python
df
```

```python
df['close'].plot()
plt.show()
df['ratio'].plot()
plt.ylabel('position ratio')
plt.show()
```

```python
df['logreturn']=np.log(df['close'])-np.log(df['close'].shift(1))
df['logreturn']
```

```python
currentPos=0
df['shortreturn']=0
for i,row in df.iterrows():
    if row['ratio']>1.15 and currentPos==0:
        currentPos=1
    elif row['ratio']>1.05 and currentPos==1:
        df.at[i,'shortreturn']=row['logreturn']
    elif row['ratio']<1.05 and currentPos==1:
        currentPos=0
currentPos=0
df['shortcumsumreturn']=df['shortreturn'].cumsum()
df['longreturn']=0
for i,row in df.iterrows():
    if row['ratio']<.85 and currentPos==0:
        currentPos=1
    elif row['ratio']<.95 and currentPos==1:
        df.at[i,'longreturn']=row['logreturn']
    elif row['ratio']>.95 and currentPos==1:
        currentPos=0
df['longcumsumreturn']=df['longreturn'].cumsum()
```

```python
df['shortcumsumreturn'].plot()
plt.ylabel('short return cumsum')
plt.show()
df['longcumsumreturn'].plot()
plt.ylabel('long return cumsum')
plt.show()
```
